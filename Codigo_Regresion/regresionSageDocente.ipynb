```python
var('a b')      # con esto, podemos usar a y b para hacer cálculo simbólico, 
                # es decir, si escribimos a+a el output será '2*a'
                # ¡Estamos usando SageMath! ¡Esto no es posible en python sin usar ninguna librería!

# El código que vamos a escribir no es el óptimo, pero familiariza al alumnado
# con el cálculo simbólico. Es como usar el papel, pero el ordenador hace las
# cuentas por ti
```

Queremos encontrar $a$ y $b$ de modo que la función $f(x) = ax+b$ minimice el error cuadrático por estos puntos.

```python
(a+b-2.1)^2    # observamos que nos escribe la expresión tal cual
```

```python
r = expand((a+b-2.1)^2)    # usando expand nos lo descompone en suma de monomios
r
```

```python
s = expand((2*a+b-3.9)^2)
s
```

```python
t = expand((3*a+b-6.2)^2)
t
```

```python
u = expand((4*a+b-7.8)^2)
u
```

```python
r+s+t+u   # este es el error
```
