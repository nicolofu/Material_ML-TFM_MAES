### Apartado 1
Encuentra el mínimo de la función $f:\mathbb{R}\to \mathbb{R}$ dada por
	$$ f(x)=e^{x^2}-2x^2-x$$

```python
e = 2.7182818284
def exp(x):
    return e**x

def f(x):
    return float(exp(x**2)-2*x**2-x)
def g(x): # derivada de f
    return float(2*x*exp(x**2)-4*x-1)
```

```python
x = 2  # empezamos con un valor
for i in range (0,200):
    print(x, f(x), g(x))
    x = x - 0.019 * g(x)  # descenso del gradiente
```

### Apartado 2
Usa el algoritmo de descenso del gradiente para hallar el mínimo de la función $g:\mathbb{R}^2\to \mathbb{R}$ definida como 
	$$g(x,y) = x^2+y^2.$$

```python
def h(x,y):
    return float(x**2 + y**2)
def grad_h(x,y):
    return [2*x, 2*y]
```

```python
x , y  = 3, 6
for i in range (0,20):
    print(x, y, h(x,y), grad_h(x,y))
    x = x - 0.3 * grad_h(x,y)[0]  # descenso del gradiente
    y = y - 0.3 * grad_h(x,y)[1]  # descenso del gradiente
```
